```python
import skimage
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

## harmony
import harmonypy as hm
from scipy.sparse.linalg import svds
import anndata as ad
import scanpy as sc
```

This script Harmony batch effect correction on the MALDI images.

Similar to modalities like scrna-seq, we can also observe batch effect of glycan signals from each slide.

Therefore we use Harmony to directly correct each pixels (same as scranseq but instead pixel == cells) glycan counts.

Similar to scRNAseq, this correct counts were not used in quantitative analysis, but instead used for community detection purposes.

This includes: 1. NN search and UMAP drawing. 2. Pixel level clustering.

```python
evalu = pd.read_csv('../extracted/EvaluationV2.csv')
#evalu = evalu.dropna()

# remove glycans that are not measured in all samples
notinlist = ['Gc_28+S', 'Gc_28+S+Na*2', 'Gc_31+Fuc+Na*2', 'Gc_37+Sx2+Na', 'Gc_48+Sx2+Na',
            'Gc_49+Fucx3', 'Gc_54+Fucx4+Na', '46', '38+Fucx2', 'Gc_8+Fuc']
evalu = evalu[~evalu['composition'].isin(notinlist)]
evalu.shape#head()#.shape
```

```python
slides = ['dsst1c1-total_ion_count', 'dsst2c2-total_ion_count',
         'dsst3c3-total_ion_count', 'dsst4c4-total_ion_count']

#slides = ['dsst1c1-total_ion_count']
root_dir = '../extracted/pipeline_results/'
glycogens = evalu.Standard_name.to_list()

containers1 = np.zeros((1182 * 3926, 38))
containers2 = np.zeros((1260 * 4092, 38))
containers3 = np.zeros((1178 * 4151, 38))
containers4 = np.zeros((1042 * 3658, 38))

####

for item in slides:
    folder_dir = '../extracted/forPixie/'+ item + '/'
    #save_dir = '../extracted/forPixie/' + item + '/'
    #os.makedirs(save_dir)
    
    idx = 0
    for glyc in glycogens:
        img_dir = folder_dir + glyc + '.tiff'
        img = skimage.io.imread(img_dir)
        
        #print(img.shape)
        fimg = img.flatten()

        
        if item == 'dsst1c1-total_ion_count':
            containers1[:,idx] = fimg
        if item == 'dsst2c2-total_ion_count':
            containers2[:,idx] = fimg
        if item == 'dsst3c3-total_ion_count':
            containers3[:,idx] = fimg
        if item == 'dsst4c4-total_ion_count':
            containers4[:,idx] = fimg
            
        idx = idx + 1
        
```

```python
container = np.vstack((containers1,containers2,containers3,containers4))
container.shape
```

```python

batch = [['1' for x in range(1182 * 3926)],
        ['2' for x in range(1260 * 4092)],
        ['3' for x in range(1178 * 4151)],
        ['4' for x in range(1042 * 3658)]]

def flatten(l):
    return [item for sublist in l for item in sublist]

batch2 = flatten(batch)
len(batch2)
```

```python
dd = {'slide':batch2, 'glycan':['g' for x in range(18497966)]}
batchdf = pd.DataFrame.from_dict(dd)
batchdf.head()
```

```python
batchdf.slide.value_counts()
```

## start harmony

### debug if all zeros too many is a probelm

```python
# take 1min
nonallzidx = ~np.all(container == 0, axis=1)
container_dez = container[nonallzidx,:]
container_dez.shape
```

```python
batchdf_dez = batchdf.iloc[nonallzidx,:]
batchdf_dez.head()
```

```python
# takes 5min
s, v, d = svds(container_dez, k = 30) # first run svd
pc = s @ np.diag(v)
```

```python
# takes ~18min
pc_hm = hm.run_harmony(pc, batchdf_dez, 'slide')
container_corrected = pc_hm.Z_corr.T @ d
container_corrected.shape
```

### check if harmony actually worked

```python
subidx = np.random.choice(3064001, 20000)
len(subidx)
```

```python
container_sub = container[subidx,:]
container_corrected_sub = container_corrected[subidx,:]
print([container_sub.shape,container_corrected_sub.shape])
```

```python
uncor_adata = ad.AnnData(
    container_sub, dtype=np.float32
)

cor_adata = ad.AnnData(
    container_corrected_sub, dtype=np.float32
)

sc.pp.scale(uncor_adata, max_value=10)
sc.pp.scale(cor_adata, max_value=10)
```

```python
sc.tl.pca(uncor_adata, svd_solver='arpack')
sc.tl.pca(cor_adata, svd_solver='arpack')
```

```python
sc.pp.neighbors(uncor_adata, n_neighbors=15, n_pcs=20)
sc.pp.neighbors(cor_adata, n_neighbors=15, n_pcs=20)
```

```python
sc.tl.umap(uncor_adata)
sc.tl.umap(cor_adata)
```

```python
uncor_adata.obs['slide'] = batchdf_dez.iloc[subidx.tolist(),0].to_numpy()
cor_adata.obs['slide'] = batchdf_dez.iloc[subidx.tolist(),0].to_numpy()
```
