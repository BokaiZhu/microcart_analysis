```python
import skimage
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
from skimage.morphology import disk, binary_erosion, binary_dilation, binary_opening, binary_closing
```

After normalizing the FOVs by dsDNA signal, the pipeline goes into two track. This track focus on the host side

In this script we remove signal aggregates (antibody) from MIBI data using empty background channels.

## below is a dialation version of aggregation removal

```python
# select two mass channel that were empty and with good range
img_dir = '../../final_data/normed/micro_fine1ms/'
img_out = '../../final_data/host_aggrmv/micro_fine1ms/'
#runs = ['2023-03-04T16-22-15-DSSC1-Run1']
#runs = ['2023-03-04T16-22-15-DSSC1-Run1', '2023-03-04T19-46-22-DSSC1-Run2'] # specific run name

#runs = ['2023-03-04T16-22-15-DSSC1-Run1', '2023-03-04T19-46-22-DSSC1-Run2',
#             '2023-03-04T23-36-59-DSSC1-Run3', '2023-03-05T15-51-42-DSST2-Run1',
#             '2023-03-05T18-22-42-DSST2-Run2', '2023-03-05T21-42-35-DSST2-Run3',
#             '2023-03-05T23-58-06-DSST2-Run4', '2023-03-06T11-23-28-DSST2-Run5',
#             '2023-03-06T13-40-46-DSST2-Run6', '2023-03-06T16-15-43-DSSC2-Run1',
#             '2023-03-07T18-32-49-DSSC2-Run5', '2023-03-07T16-08-39-DSSC2-Run4',
#             '2023-03-07T13-54-48-DSSC2-Run3', '2023-03-07T11-17-56-DSSC2-Run2',
#             '2023-03-09T15-53-36-DSSC3-Run4', '2023-03-09T13-34-52-DSSC3-Run3',
#             '2023-03-09T10-32-36-DSSC3-Run2', '2023-03-08T23-51-06-DSSC3-Run1',
#             '2023-03-08T21-26-09-DSST3-Run6', '2023-03-08T18-48-43-DSST3-Run5',
#             '2023-03-08T16-15-19-DSST3-Run4', '2023-03-08T14-03-50-DSST3-Run3',
#             '2023-03-08T11-39-28-DSST3-Run2', '2023-03-07T22-15-40-DSST3-Run1'] # specific run name

runs = ['2023-03-10T12-31-44-DSST4-Run1', '2023-03-10T14-47-38-DSST4-Run2', '2023-03-10T17-34-14-DSST4-Run3',
            '2023-03-18T14-36-10-DSST4-Run4', '2023-03-18T17-28-03-DSST4-Run5', '2023-03-18T23-11-36-DSST4-Run6',
            '2023-03-19T12-45-44-DSSC4-Run1', '2023-03-19T15-20-40-DSSC4-Run2', '2023-03-19T17-38-15-DSSC4-Run3',
            '2023-03-19T20-32-23-DSSC4-Run4', '2023-03-20T12-07-46-DSST1-Run1', '2023-03-20T14-32-22-DSST1-Run2',
            '2023-03-20T19-13-53-DSST1-Run3', '2023-03-20T21-44-51-DSST1-Run4', '2023-03-21T24-29-43-DSST1-Run5',
            '2023-03-21T10-32-42-DSST1-Run6', '2023-03-21T12-44-51-DSST1-Run7', '2023-03-28T11-45-15-DSSC1-Run4'
            ]

empty = ['chan_163'] # good empty channels to detect potential aggregation

for run in runs:
    
    path = os.path.join(img_dir, run)
    fovs = os.listdir(path)
    try:
        fovs.remove('.DS_Store')
    except:
        pass
    
    for fov in fovs:
        print(fov + run)
        mask = np.zeros((1024,1024)) # standard fov setting
        imgs = os.listdir(img_dir + run + '/' + fov + '/')
        os.makedirs(img_out + run + '/' + fov + '/', exist_ok=True)
    
        for img in empty:
            dir_crr = img_dir + run + '/' + fov + '/' + img + '.tiff'
            tif = skimage.io.imread(dir_crr)

            qq = 0.04 # qq designed for channel 163
            temp = tif.copy()
            temp[temp>qq] = 1
            mask = temp.copy()
            #plt.imshow(mask)
            #plt.show()

        ## turn mask into binary
        mask[mask>=1] = 1
        mask[mask<1] = 0
        
        dilated = binary_dilation(mask, disk(4))
        # save out agg_removed version
        for channel in imgs:
            dir_crr = img_dir + run + '/' + fov + '/' + channel
            cc = skimage.io.imread(dir_crr)
            cc2 = cc * (1-dilated)

            sv_dir = img_out + run + '/' + fov + '/'  + channel
            skimage.io.imsave(sv_dir, cc2)
```

## below this is the previous version that is not aggressive enough

```python
# select two mass channel that were empty and with good range
img_dir = '../../final_data/normed/micro_fine1ms/'
img_out = '../../final_data/host_aggrmv/micro_fine1ms/'
#runs = ['2023-03-04T16-22-15-DSSC1-Run1']
runs = ['2023-03-04T16-22-15-DSSC1-Run1', '2023-03-04T19-46-22-DSSC1-Run2',
             '2023-03-04T23-36-59-DSSC1-Run3', '2023-03-05T15-51-42-DSST2-Run1',
             '2023-03-05T18-22-42-DSST2-Run2', '2023-03-05T21-42-35-DSST2-Run3',
             '2023-03-05T23-58-06-DSST2-Run4', '2023-03-06T11-23-28-DSST2-Run5',
             '2023-03-06T13-40-46-DSST2-Run6', '2023-03-06T16-15-43-DSSC2-Run1',
             '2023-03-07T18-32-49-DSSC2-Run5', '2023-03-07T16-08-39-DSSC2-Run4',
             '2023-03-07T13-54-48-DSSC2-Run3', '2023-03-07T11-17-56-DSSC2-Run2',
             '2023-03-09T15-53-36-DSSC3-Run4', '2023-03-09T13-34-52-DSSC3-Run3',
             '2023-03-09T10-32-36-DSSC3-Run2', '2023-03-08T23-51-06-DSSC3-Run1',
             '2023-03-08T21-26-09-DSST3-Run6', '2023-03-08T18-48-43-DSST3-Run5',
             '2023-03-08T16-15-19-DSST3-Run4', '2023-03-08T14-03-50-DSST3-Run3',
             '2023-03-08T11-39-28-DSST3-Run2', '2023-03-07T22-15-40-DSST3-Run1'] # specific run name

empty = ['chan_163','chan_140'] # good empty channels to detect potential aggregation
mask = np.zeros((1024,1024)) # standard fov setting

for run in runs:
    
    path = os.path.join(img_dir, run)
    fovs = os.listdir(path)
    try:
        fovs.remove('.DS_Store')
    except:
        pass
    
    for fov in fovs:
        
        imgs = os.listdir(img_dir + run + '/' + fov + '/')
        os.makedirs(img_out + run + '/' + fov + '/', exist_ok=True)
    
        for img in empty:
            dir_crr = img_dir + run + '/' + fov + '/' + img + '.tiff'
            tif = skimage.io.imread(dir_crr)

            # quantile threshold
            #qq = np.quantile(tif, 0.9997) # this value should be good, no more testing needed
            # after testing a hard coded threshold should work better
            qq = 0.06
            temp = tif.copy()
            temp[temp>qq] = 1

            # gaussian filter
            gau_temp = skimage.filters.gaussian(temp, sigma=2, truncate=2)
            #qq2  = np.quantile(gau_temp,0.995) # this value should be good, no more testing needed
            qq2  = np.quantile(gau_temp,0.9999) # this value should be good, no more testing needed
            tt2 = gau_temp.copy()
            tt2[gau_temp>qq2] = 1
            mask = mask + tt2

        ## turn mask into binary
        mask[mask>=1] = 1
        mask[mask<1] = 0

        # save out agg_removed version
        for channel in imgs:
            dir_crr = img_dir + run + '/' + fov + '/' + channel
            cc = skimage.io.imread(dir_crr)
            cc2 = cc * (1-mask)

            sv_dir = img_out + run + '/' + fov + '/'  + channel
            skimage.io.imsave(sv_dir, cc2)
```
