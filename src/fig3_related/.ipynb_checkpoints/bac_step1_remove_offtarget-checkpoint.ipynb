```python
# create csv of single cell extracted

img_dir = '../../final_data/normed/micro_fine1ms/'

#runs = ['2023-03-04T16-22-15-DSSC1-Run1', '2023-03-04T19-46-22-DSSC1-Run2',
#             '2023-03-04T23-36-59-DSSC1-Run3', '2023-03-05T15-51-42-DSST2-Run1',
#             '2023-03-05T18-22-42-DSST2-Run2', '2023-03-05T21-42-35-DSST2-Run3',
#             '2023-03-05T23-58-06-DSST2-Run4', '2023-03-06T11-23-28-DSST2-Run5',
#             '2023-03-06T13-40-46-DSST2-Run6', '2023-03-06T16-15-43-DSSC2-Run1',
#             '2023-03-07T18-32-49-DSSC2-Run5', '2023-03-07T16-08-39-DSSC2-Run4',
#             '2023-03-07T13-54-48-DSSC2-Run3', '2023-03-07T11-17-56-DSSC2-Run2',
#             '2023-03-09T15-53-36-DSSC3-Run4', '2023-03-09T13-34-52-DSSC3-Run3',
#             '2023-03-09T10-32-36-DSSC3-Run2', '2023-03-08T23-51-06-DSSC3-Run1',
#             '2023-03-08T21-26-09-DSST3-Run6', '2023-03-08T18-48-43-DSST3-Run5',
#             '2023-03-08T16-15-19-DSST3-Run4', '2023-03-08T14-03-50-DSST3-Run3',
#             '2023-03-08T11-39-28-DSST3-Run2', '2023-03-07T22-15-40-DSST3-Run1'] # specific run name

#runs = ['2023-03-10T12-31-44-DSST4-Run1', '2023-03-10T14-47-38-DSST4-Run2', '2023-03-10T17-34-14-DSST4-Run3',
#            '2023-03-18T14-36-10-DSST4-Run4', '2023-03-18T17-28-03-DSST4-Run5', '2023-03-18T23-11-36-DSST4-Run6',
#            '2023-03-19T12-45-44-DSSC4-Run1', '2023-03-19T15-20-40-DSSC4-Run2', '2023-03-19T17-38-15-DSSC4-Run3',
#            '2023-03-19T20-32-23-DSSC4-Run4', '2023-03-20T12-07-46-DSST1-Run1', '2023-03-20T14-32-22-DSST1-Run2',
#            '2023-03-20T19-13-53-DSST1-Run3', '2023-03-20T21-44-51-DSST1-Run4', '2023-03-21T24-29-43-DSST1-Run5',
#            '2023-03-21T10-32-42-DSST1-Run6', '2023-03-21T12-44-51-DSST1-Run7', '2023-03-28T11-45-15-DSSC1-Run4'
#            ]

runs = ['2023-03-10T12-31-44-DSST4-Run1', '2023-03-10T14-47-38-DSST4-Run2', '2023-03-10T17-34-14-DSST4-Run3',
            '2023-03-18T14-36-10-DSST4-Run4', '2023-03-18T17-28-03-DSST4-Run5', '2023-03-18T23-11-36-DSST4-Run6',
            '2023-03-19T12-45-44-DSSC4-Run1', '2023-03-19T15-20-40-DSSC4-Run2', '2023-03-19T17-38-15-DSSC4-Run3',
            '2023-03-19T20-32-23-DSSC4-Run4', '2023-03-20T12-07-46-DSST1-Run1', '2023-03-20T14-32-22-DSST1-Run2',
            '2023-03-20T19-13-53-DSST1-Run3', '2023-03-20T21-44-51-DSST1-Run4', '2023-03-21T24-29-43-DSST1-Run5',
            '2023-03-21T10-32-42-DSST1-Run6', '2023-03-21T12-44-51-DSST1-Run7', '2023-03-28T11-45-15-DSSC1-Run4',
            '2023-03-04T16-22-15-DSSC1-Run1', '2023-03-04T19-46-22-DSSC1-Run2',
             '2023-03-04T23-36-59-DSSC1-Run3', '2023-03-05T15-51-42-DSST2-Run1',
             '2023-03-05T18-22-42-DSST2-Run2', '2023-03-05T21-42-35-DSST2-Run3',
             '2023-03-05T23-58-06-DSST2-Run4', '2023-03-06T11-23-28-DSST2-Run5',
             '2023-03-06T13-40-46-DSST2-Run6', '2023-03-06T16-15-43-DSSC2-Run1',
             '2023-03-07T18-32-49-DSSC2-Run5', '2023-03-07T16-08-39-DSSC2-Run4',
             '2023-03-07T13-54-48-DSSC2-Run3', '2023-03-07T11-17-56-DSSC2-Run2',
             '2023-03-09T15-53-36-DSSC3-Run4', '2023-03-09T13-34-52-DSSC3-Run3',
             '2023-03-09T10-32-36-DSSC3-Run2', '2023-03-08T23-51-06-DSSC3-Run1',
             '2023-03-08T21-26-09-DSST3-Run6', '2023-03-08T18-48-43-DSST3-Run5',
             '2023-03-08T16-15-19-DSST3-Run4', '2023-03-08T14-03-50-DSST3-Run3',
             '2023-03-08T11-39-28-DSST3-Run2', '2023-03-07T22-15-40-DSST3-Run1']

listm = []

out_container = []
for run in runs:
    print("working on ", run)
    container = []
    path_masked = os.path.join(img_dir, run)
    fovs = os.listdir(path_masked)
    
    for fov in fovs:
        
        if fov == '.DS_Store':
            continue
        
        ####### remove off target signal for 488
        
        ## 488 off signal caused by large aggregation
        alx488 = skimage.io.imread( img_dir + run + '/' + fov + '/' + 'Alx488.tiff')
        qq = np.quantile(alx488, 0.99) # 0.995?
        temp = alx488.copy()
        temp[temp>qq] = qq
        
        tt = mh.thresholding.bernsen(temp, 10, 15)
        ttm = skimage.measure.label(~tt)
        
        rmvbig = skimage.morphology.remove_small_objects(ttm, 100) # objects larger than 100 pixels
        rmvbig[rmvbig>0]=1
        
        dilated = binary_dilation(rmvbig, disk(3)) # 2?
        alx488v2 = alx488 * (1-dilated)
        
        #skimage.io.imsave(img_dir + run + '/' + fov + '/' + 'Alx488_v2.tiff', alx488v2)
        ######## remove 488 off target finished
        
        ######## start to remove off target caused by mucus singal capturing
        
        muc = skimage.io.imread( img_dir + run + '/' + fov + '/' + 'MUC2.tiff')
        dig = skimage.io.imread( img_dir + run + '/' + fov + '/' + 'DIG.tiff')
        tam = skimage.io.imread( img_dir + run + '/' + fov + '/' + 'TAMRA.tiff')
        ftc = skimage.io.imread( img_dir + run + '/' + fov + '/' + 'FITC.tiff')
        
        #### remove signal bleedthrough from muc
        dig = dig - muc * 0.05
        dig[dig<0] = 0
        tam = tam - muc * 0.05
        tam[tam<0] = 0
        ftc = ftc - muc * 0.05
        ftc[ftc<0] = 0
        alx488v2 = alx488v2 - muc * 0.05
        alx488v2[alx488v2<0] = 0
        #########################################
        
        ### then remove bg noise specific to empty imaging of gold on fecal area
        
        alx488v2[alx488v2<=0.005] = 0
        ftc[ftc<=0.003] = 0
        dig[dig<=0.003] = 0
        tam[tam<=0.003] = 0
        
        ## also scale up 1000 fold for better img viz
        #alx488v2 = alx488v2 * 100
        #ftc = ftc * 100
        #dig = dig * 100
        #tam = tam * 100
        
        skimage.io.imsave(img_dir + run + '/' + fov + '/' + 'DIG_v3.tiff', dig)
        skimage.io.imsave(img_dir + run + '/' + fov + '/' + 'TAMRA_v3.tiff', tam)
        skimage.io.imsave(img_dir + run + '/' + fov + '/' + 'FITC_v3.tiff', ftc)
        skimage.io.imsave(img_dir + run + '/' + fov + '/' + 'Alx488_v3.tiff', alx488v2)
```
