```python
import skimage
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')

# packages used for this script specifically
from skimage import filters
from skimage import exposure
import mahotas as mh
from skimage.morphology import disk, binary_erosion, binary_dilation, binary_opening, binary_closing

from itertools import compress
from scipy import stats
```

This script calculates a localized bacteria diversity entropy based on MIBI imaging data.

```python
# some helper function: # define a quick function for sliding windon

def sliding_window(image, stepSize, windowSize):
    for y in range(50, image.shape[0], stepSize):
        for x in range(50, image.shape[1], stepSize):
            yield (x, y, image[y:y + windowSize[1], x:x + windowSize[0]])
        
# define action function:
def extractFeatures(window):
    total = window[2].shape[0] * window[2].shape[1]
    sigsum = np.sum(window[2])
    perc = sigsum / total
    return perc

# define binarize function

def binarize(image, thresh):
    temp = image.copy()
    temp[temp>=thresh] = 1 # these are hard coded values that has been manually evaluated
    temp[temp<thresh] = 0
    return temp
```

```python
# create csv of single cell extracted


csv_out = '../../../final_data/bac_csv/'
img_dir = '../../../final_data/normed/micro_fine1ms/'
img_dir_mask = '../../../final_data/masked/micro_fine1ms/'


#runs = ['2023-03-04T23-36-59-DSSC1-Run3']

runs = ['2023-03-04T16-22-15-DSSC1-Run1', '2023-03-04T19-46-22-DSSC1-Run2',
             '2023-03-04T23-36-59-DSSC1-Run3', '2023-03-05T15-51-42-DSST2-Run1',
             '2023-03-05T18-22-42-DSST2-Run2', '2023-03-05T21-42-35-DSST2-Run3',
             '2023-03-05T23-58-06-DSST2-Run4', '2023-03-06T11-23-28-DSST2-Run5',
             '2023-03-06T13-40-46-DSST2-Run6', '2023-03-06T16-15-43-DSSC2-Run1',
             '2023-03-07T18-32-49-DSSC2-Run5', '2023-03-07T16-08-39-DSSC2-Run4',
             '2023-03-07T13-54-48-DSSC2-Run3', '2023-03-07T11-17-56-DSSC2-Run2',
             '2023-03-09T15-53-36-DSSC3-Run4', '2023-03-09T13-34-52-DSSC3-Run3',
             '2023-03-09T10-32-36-DSSC3-Run2', '2023-03-08T23-51-06-DSSC3-Run1',
             '2023-03-08T21-26-09-DSST3-Run6', '2023-03-08T18-48-43-DSST3-Run5',
             '2023-03-08T16-15-19-DSST3-Run4', '2023-03-08T14-03-50-DSST3-Run3',
             '2023-03-08T11-39-28-DSST3-Run2', '2023-03-07T22-15-40-DSST3-Run1',
            '2023-03-10T12-31-44-DSST4-Run1', '2023-03-10T14-47-38-DSST4-Run2', '2023-03-10T17-34-14-DSST4-Run3',
            '2023-03-18T14-36-10-DSST4-Run4', '2023-03-18T17-28-03-DSST4-Run5', '2023-03-18T23-11-36-DSST4-Run6',
            '2023-03-19T12-45-44-DSSC4-Run1', '2023-03-19T15-20-40-DSSC4-Run2', '2023-03-19T17-38-15-DSSC4-Run3',
            '2023-03-19T20-32-23-DSSC4-Run4', '2023-03-20T12-07-46-DSST1-Run1', '2023-03-20T14-32-22-DSST1-Run2',
            '2023-03-20T19-13-53-DSST1-Run3', '2023-03-20T21-44-51-DSST1-Run4', '2023-03-21T24-29-43-DSST1-Run5',
            '2023-03-21T10-32-42-DSST1-Run6', '2023-03-21T12-44-51-DSST1-Run7', '2023-03-28T11-45-15-DSSC1-Run4'
       ]


####### start running

fovll = []
runll = []
entroll = []

for run in runs:
    print("working on ", run)
    path_masked = os.path.join(img_dir, run)
    fovs = os.listdir(path_masked)
    
    for fov in fovs:
        
        if fov == '.DS_Store':
            continue
        
        if os.path.exists(img_dir_mask + run + '/' + fov + '/Mask.tif') == False: # fov without mask no bacteria
            continue
        
        ## read in the nec images
        alx488 = skimage.io.imread( img_dir + run + '/' + fov + '/Alx488_v3.tiff')
        tam = skimage.io.imread( img_dir + run + '/' + fov + '/TAMRA_v3.tiff')
        dig = skimage.io.imread( img_dir + run + '/' + fov + '/DIG_v3.tiff')
        fitc = skimage.io.imread( img_dir + run + '/' + fov + '/FITC_v3.tiff')
        
        # preprocess that binarize, capping first
        q2 = np.quantile(alx488, 0.001)
        alx488[alx488<q2] = q2
        tam[tam<q2] = q2
        dig[dig<q2] = q2
        fitc[fitc<q2] = q2
        
        # binarize
        alx488_b = binarize(alx488, 0.015)
        tam_b = binarize(tam, 0.01)
        dig_b = binarize(dig, 0.01)
        fitc_b = binarize(fitc, 0.01)
        
        ########## start calculating correlations
        #plt.imshow(alx488_b)
        #plt.show()
        
        #plt.imshow(tam_b)
        #plt.show()
        
        #plt.imshow(dig_b)
        #plt.show()
        
        #plt.imshow(fitc_b)
        #plt.show()
        
        ########## start to calculate entrophy
        
        ### for 488
        windows_488 = sliding_window(alx488_b, 10, (100, 100)) # slide step 1, window size 100 pix
        perc_list_488 = []

        for window in windows_488:
            featureVector = extractFeatures(window)
            perc_list_488.append(featureVector)
            
        ### for dig
        windows_dig = sliding_window(dig_b, 10, (100, 100)) # slide step 1, window size 100 pix
        perc_list_dig = []

        for window in windows_dig:
            featureVector = extractFeatures(window)
            perc_list_dig.append(featureVector)
            
        ### for tam
        windows_tam = sliding_window(tam_b, 10, (100, 100)) # slide step 1, window size 100 pix
        perc_list_tam = []

        for window in windows_tam:
            featureVector = extractFeatures(window)
            perc_list_tam.append(featureVector)
            
        ### for fitc
        windows_fitc = sliding_window(fitc_b, 10, (100, 100)) # slide step 1, window size 100 pix
        perc_list_fitc = []

        for window in windows_fitc:
            featureVector = extractFeatures(window)
            perc_list_fitc.append(featureVector)
        
        ### remove windows that are too empty:
        empid = [x != 0 for x in perc_list_488]
        perc_list_488 = list(compress(perc_list_488, empid))
        perc_list_dig = list(compress(perc_list_dig, empid))
        perc_list_tam = list(compress(perc_list_tam, empid))
        perc_list_fitc = list(compress(perc_list_fitc, empid))
        
        ### calc entro
        entrollfov = []
        
        for i in range(len(perc_list_fitc)):
            entro = stats.entropy(pk= [perc_list_488[i],
                                      perc_list_dig[i],
                                      perc_list_tam[i],
                                      perc_list_fitc[i]],base=2)
            entrollfov.append(entro)
        
        ### save the results
        fovll.append(fov)
        runll.append(run)
        entroll.append(np.median(entrollfov))
        
```

```python
dd = {'fov':fovll, 'run':runll, 'entro':entroll}

data = pd.DataFrame.from_dict(dd)
data
```

```python
data.to_csv('../../../final_data/bac_csv/nbhd_entroV1.csv')
```
