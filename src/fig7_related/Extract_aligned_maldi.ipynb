```python
import skimage
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

quick code to extract masked regions from  glycan images.

masked regions were created manually, to align regions from the three different modalities.

```python
# prep result container
dd = {}
a = pd.read_csv('../../maldi/Extracted_sigNormsize_all.csv')
glycs = a.columns[2:].tolist()

for g in glycs:
    dd[g] = []
dd['id'] = []
#dd
```

```python
slides = ['dsst1c1-total_ion_count', 'dsst2c2-total_ion_count',
          'dsst3c3-total_ion_count','dsst4c4-total_ion_count']

#slides = ['dsst1c1-total_ion_count']

for slide in slides:
    dir = '../../maldi/extracted/pipeline_results/' + slide 
    mask_dir = dir + '/Mask*.tif'
    masklist = glob.glob(mask_dir)
    masklist

    # create a slide full maks
    contain = np.zeros_like(skimage.io.imread(masklist[0]))
    for msk in masklist:
        msk = skimage.io.imread(msk)
        msk[msk > 0] = 1
        contain = contain + msk

    # find all fovs
    fov_dir = dir + '/match/*.tif'
    fovlist = glob.glob(fov_dir)

    ## loop through all channels
    gly_dir = '../../maldi/extracted/forPixie/' + slide 
    # then get each individual fov
    img_dir = gly_dir + '/*.tiff'
    imglist = glob.glob(img_dir)

    i = 0
    for glc in imglist: # for each channel
        gimg = skimage.io.imread(glc)# get image
        gname = glc.split('/')[-1][:-5]
        
        for fov in fovlist: # for each fov
            if i == 0:
                #print(int(fov.split('/')[-1][:-4]))
                dd['id'].append(int(fov.split('/')[-1][:-4]))

            # get image
            fov = skimage.io.imread(fov)
            fov[fov > 0] = 1
            fov = np.rot90(fov, axes = (1,0)) # rotate back
            f_msk = fov * contain
            # extract
            res = np.sum(gimg * f_msk) / np.sum(f_msk) # value scaled by area incase that matters
            dd[gname].append(res)
        i = i + 1
                
```

```python
fov_maldi = pd.DataFrame.from_dict(dd)
fov_maldi
```

```python
fov_maldi.to_csv('../data/fov_maldi_extracted.csv', index = False)
```
